```python
import numpy as np
import multiprocessing as mp
from time import time
```

```python
# Prepare data
np.random.RandomState(100) 
arr = np.random.randint(0, 10, size=[30000, 30000]) #NumPy 生成一个大型随机整数二维数组
data = arr.tolist()
#data[:5]
```

```python
def examp01(row, minimum, maximum):
    count = 0
    for n in row:
        if minimum <= n <= maximum:
            count = count + 1
    return count
```


### Serial Execution


对 data 中的每一行数据，串行（逐行）执行 examp01 函数，并记录总耗时。

```python
results = []

t1 = time()
for row in data:
    results.append(examp01(row, minimum=4, maximum=8))

print('Elapsed time: \n', time() - t1)
    
print('Serial execution:', results[:10])
```


### Pool map()


```python
def examp01_rowonly(row, minimum=4, maximum=8):
    count = 0
    for n in row:
        if minimum <= n <= maximum:
            count = count + 1
    return count
```

```python
t1 = time()
pool = mp.Pool(2)

results = pool.map(examp01_rowonly, [row for row in data])

pool.close()
print('Elapsed time: \n', time() - t1)
print('Pool.map execution:',results[:10])
```

```python
t1 = time()
pool = mp.Pool(4)

results = pool.map(examp01_rowonly, [row for row in data])

pool.close()
print('Elapsed time: \n', time() - t1)
print('Pool.map execution:',results[:10])
```


### Pool starmap()


```python
t1 = time()
pool = mp.Pool(4)

results = pool.starmap_async(examp01, [(row, 4, 8) for row in data])

pool.close()
print('Elapsed time: \n', time() - t1)
print('Pool.starmap execution:', results[:10])
```


### Pool imap()


```python
t1 = time()
pool = mp.Pool(4)

result = pool.imap(examp01_rowonly, [row for row in data],50)
results = [x for x in result]

pool.close()
print('Elapsed time: \n', time() - t1)
print('Pool.map execution:',results[:10])
```
